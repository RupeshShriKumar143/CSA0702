```python
import pandas as pd

# Loan Approval Dataset
data = {
    'Income': ['High', 'High', 'Low', 'High'],
    'CreditScore': ['Good', 'Good', 'Bad', 'Bad'],
    'Collateral': ['Yes', 'No', 'No', 'Yes'],
    'Approved': ['Yes', 'Yes', 'No', 'Yes']
}

df = pd.DataFrame(data)

print("Dataset:\n")
print(df)
```

```python
import copy

# -------------------------------
# Candidate Elimination Algorithm
# -------------------------------

concepts = df.iloc[:, :-1].values
target = df.iloc[:, -1].values

num_attributes = len(concepts[0])

# Initialize S and G
S = ['Ø'] * num_attributes
G = [['?'] * num_attributes]

print("\nInitial Specific Boundary (S):", S)
print("Initial General Boundary (G):", G)

for i, example in enumerate(concepts):

    print("\n===================================")
    print("Processing Example", i + 1)
    print("Example:", example)
    print("Class:", target[i])

    # -----------------------------
    # Positive Example
    # -----------------------------
    if target[i] == "Yes":

        # First positive example
        if S[0] == 'Ø':
            S = list(example)

        else:
            for j in range(num_attributes):
                if S[j] != example[j]:
                    S[j] = '?'

        # Remove inconsistent hypotheses from G
        G = [g for g in G if all(
            g[k] == '?' or g[k] == example[k]
            for k in range(num_attributes)
        )]

    # -----------------------------
    # Negative Example
    # -----------------------------
    else:

        new_G = []

        for g in G:

            for j in range(num_attributes):

                if g[j] == '?':

                    if S[j] != '?' and S[j] != 'Ø':

                        new_hypothesis = g.copy()
                        new_hypothesis[j] = S[j]

                        # Reject negative example
                        if new_hypothesis[j] != example[j]:
                            new_G.append(new_hypothesis)

        G = new_G

    print("Specific Boundary (S):", S)
    print("General Boundary (G):", G)

print("\n===================================")
print("Final Specific Boundary (S):", S)
print("Final General Boundary (G):", G)
```

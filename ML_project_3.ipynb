```python
# -----------------------------------------
# General-or-Specific? (Partial Order Reasoning)
# -----------------------------------------

def compare_hypotheses(h1, h2):

    more_general = False
    more_specific = False

    for a, b in zip(h1, h2):

        if a == b:
            continue

        elif a == '?':
            more_general = True

        elif b == '?':
            more_specific = True

        elif a == 'Ø':
            more_specific = True

        elif b == 'Ø':
            more_general = True

        else:
            return "Not Comparable"

    if more_general and not more_specific:
        return "h1 is more general than h2"

    elif more_specific and not more_general:
        return "h2 is more general than h1"

    elif not more_general and not more_specific:
        return "Both are equally general"

    else:
        return "Not Comparable"


# -------------------------------
# Questions
# -------------------------------

questions = [

    (['?','Warm','?'],
     ['Sunny','Warm','?']),

    (['Sunny','?','Normal'],
     ['?','Warm','?']),

    (['Ø','Warm','?'],
     ['Sunny','Warm','Normal']),

    (['?','?','?'],
     ['?','?','?']),

    (['Sunny','Cold','?'],
     ['Sunny','?','High'])
]

# -------------------------------
# Display Results
# -------------------------------

for i, (h1, h2) in enumerate(questions, start=1):

    print("="*60)
    print("Question", i)
    print("h1 =", h1)
    print("h2 =", h2)

    result = compare_hypotheses(h1, h2)

    print("Answer:", result)
```
